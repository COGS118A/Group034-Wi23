# Evaluation Metrics

For evaluation, we will utilize confusion matrix to intuitively analyze the discrepancy between our models’ predictions and the ground truth. The confusion matrix will be of shape 2x2 and cells containing true positives, false positives, false negatives and true negatives. We will also take advantage of the AUC-ROC curve to make sure our models have good classification threshold. We can then fine tune our models to maximize specificity / precision based on our demand. Another evaluation metric will be prediction accuracy, which is the ratio of correct predictions out of all predictions. All the models will be tested on the same test dataset that is randomly splitted from the shuffled full dataset before training. The size of test data will be 20% of all data and the rest of the data will be used for training. We picked this ratio considering we have sufficient data. Thus, a large test dataset will improve our model’s robustness.

Precision is the ratio of true positive predictions to the total positive predictions, as shown below:

Precision = True Positives / (True Positives + False Positives)

Specificity is the ratio of true negative predictions to the total negative predictions, as shown below:

Specificity = True Negatives / (True Negatives + False Positives)

We want to fine-tune the models for these metrics because they are important in the context of our problem. In the case of our problem, a high precision means that the model correctly identifies clients who are likely to subscribe to a term deposit, which can help the marketing team to focus their efforts on these clients. On the other hand, a high specificity means that the model correctly identifies clients who are unlikely to subscribe to a term deposit, which can help the marketing team to avoid wasting resources.

We will use the confusion matrix to calculate precision and specificity for each model. The confusion matrix shows the true positive, false positive, false negative, and true negative values for the model's predictions. The AUC-ROC curve will also be used to ensure that the models have a good classification threshold.

In addition to precision and specificity, we will also use prediction accuracy to evaluate the performance of our models. Prediction accuracy is the ratio of correct predictions to the total number of predictions. We will randomly split our dataset into training and testing sets, with 20% of the data reserved for testing. This split ratio is appropriate for our dataset size, and using a large test dataset will improve the robustness of our model's performance.

### Subsection 2: Data Engineering

First, we make a hard copy of the dataset and call it 'bank_data_modified'.

We assign +1 for 'yes' and -1 for 'no' in the 'y' column, and then split the column into a new nparray variable called 'y'.

```python
bank_data_modified = bank_data.copy()
bank_data_modified.loc[bank_data_modified['y']=='yes','y'] = 1
bank_data_modified.loc[bank_data_modified['y']=='no','y'] = -1
bank_data_modified['y'] = bank_data_modified['y'].astype(int)
bank_data_modified
```

This is what 'y' looks like:

```python
y = bank_data_modified.pop('y')
y
```

Then, we pre-process the '-1' values in the 'pdays' column as stated in "Data" section above, and apply a normalization, in this case, a MinMaxSacler(), which normalize all values to be between 0 and 1, to all the numerical columns

```python
pdays_max_value = bank_data_modified['pdays'].max()
bank_data_modified.loc[bank_data_modified['pdays']==-1, 'pdays'] = pdays_max_value
num_cols = bank_data_modified.select_dtypes(include='number').columns
scaler = MinMaxScaler()
for col in num_cols:
    bank_data_modified[col] = scaler.fit_transform(bank_data_modified[[col]])
bank_data_modified
```

Then, we apply a one-hot encoding with an exception of 'unknown' values that are coded as a soft one-hot with 1/n across all n dimensions of a feature, as stated in the "Data" section, to all the categorical columns.

```python
cat_cols = bank_data_modified.select_dtypes(include='object').columns
for col in cat_cols:
    one_hot = pd.get_dummies(bank_data[col], prefix=col, )
    mask = (bank_data_modified[col] == 'unknown')
    n = len(bank_data_modified[col].unique()) - 1 #exclude 'unknown'
    one_hot.loc[mask, :] = 1/n
    if (col+'_unknown') in one_hot:
        one_hot.drop((col+'_unknown'), axis=1, inplace=True)
    bank_data_modified = pd.concat([bank_data_modified, one_hot], axis=1)
    bank_data_modified.drop(columns=[col], inplace=True)
```

Last, we rename the dataframe to be 'X', which is conventional for the learning process.

This is what 'X' looks like:

```python
X = bank_data_modified
X
```

### Limitations

Although all three of our models have pretty good accuracy, we consider them to have poor performance. In real-life sceanario, the most important job for the banks are to detect potential customers who will make a deposit, which means that Recall is very important for a prediction model; however, all three of our models have very poor Recall. This could be due to the original dataset. The dataset has overwhelmingly more negative cases than positive ones (in our test data, 7952 to 1091), which means that a predictor that gives negative predictions all the time would have a accuracy at around 0.88, which is very close to the results we get. Thus, we infer that it is the unbalanced portion of positive and negative cases in the original dataset that leads our models to be very prone to giving neagative predictions. A possible solution could be that, with more data, we can randomly pick a subset of the original dataset, which contains about the same positive and negative cases, and train the model on that subset.

Another problem of the original dataset is that the dataset itself is very noisy, not in the sense of missing or wrong values, but in the sense of that many features in the dataset are not contributing to the prediction (e.g., pdays is a very week predictor comparing to job), and in the sense that numerical datas vary too much in their scales (e.g., for the balance feature, most variables become less than 0.1 after normalization. A possible solution to this problem is that we could do more EDA on the original dataset to cut off insignificant features, and apply a none-linear normalization on numerical numbers to make the values less distant in scales.
